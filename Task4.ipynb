```python
import statistics

def summarize(values):
    summary = {
        "count": len(values),
        "mean": statistics.mean(values),
        "min": min(values),
        "max": max(values),
        "std_dev": statistics.stdev(values)
    }
    return summary

# Example usage
data = [12, 45, 67, 23, 89, 34, 56, 78, 90, 11, 29, 38, 50, 72, 81]
result = summarize(data)
print(result)
```

```python
#  Linear Regression on Stress Dataset
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("/content/Stress_Dataset.csv")

# Select numeric features and target
X = df[[
    "Have you recently experienced stress in your life?",
    "Do you face any sleep problems or difficulties falling asleep?",
    "Have you been getting headaches more often than usual?"
]]
y = df["Age"]

# Initialize and train the model
model = LinearRegression()
model.fit(X, y)

# Print coefficients and intercept
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

```python
import pandas as pd

df = pd.read_csv("/content/Stress_Dataset.csv")

# Display column names
print(df.columns)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convert Yes/No to numeric (apply to whole DataFrame)
df_encoded = df.copy()
df_encoded = df_encoded.applymap(lambda x: 1 if str(x).lower() == 'yes' else (0 if str(x).lower() == 'no' else x))

# Keep only numeric columns
numeric_df = df_encoded.select_dtypes(include=['int64', 'float64'])

# Correlation matrix
corr = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Heatmap of Numerical Features", fontsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Convert Yes/No to numeric for one survey question
df['stress_experience'] = df['Have you recently experienced stress in your life?'].map({'Yes':1, 'No':0})

# Scatter Age vs Stress experience
plt.figure(figsize=(8,6))
plt.scatter(df['Age'], df['stress_experience'], alpha=0.6, edgecolor="k")
plt.xlabel("Age")
plt.ylabel("Stress Experience (1=Yes, 0=No)")
plt.title("Scatterplot: Age vs Stress Experience")
plt.show()
```

```python
# Make a copy of dataset
df_cleaned = df.copy()

# Convert Yes/No answers to numeric (1/0)
df_cleaned = df_cleaned.applymap(lambda x: 1 if str(x).lower() == 'yes' else (0 if str(x).lower() == 'no' else x))

# Handle missing values (fill numeric NaN with mean, categorical with mode)
for col in df_cleaned.columns:
    if df_cleaned[col].dtype in ['int64', 'float64']:
        df_cleaned[col].fillna(df_cleaned[col].mean(), inplace=True)
    else:
        df_cleaned[col].fillna(df_cleaned[col].mode()[0], inplace=True)

# Save cleaned dataset
df_cleaned.to_csv("cleaned_dataset.csv", index=False)

print(" Cleaned dataset saved as 'cleaned_dataset.csv'")
```
