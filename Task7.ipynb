```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
column_names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
iris = pd.read_csv("iris.data", header=None, names=column_names)

# Extract Petal Length column
X = iris['petal_length']

# Sort values for CDF
sorted_X = np.sort(X)
cdf = np.arange(1, len(sorted_X)+1) / len(sorted_X)

# Plot CDF
plt.figure(figsize=(8,5))
plt.plot(sorted_X, cdf, marker='.', linestyle='-', color="blue")

plt.title("Cumulative Distribution Function (CDF) of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Cumulative Probability")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import gaussian_kde

# Load dataset
column_names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
iris = pd.read_csv("iris.data", header=None, names=column_names)

# Extract Sepal Width column
X = iris['sepal_width']

# --- PDF using Kernel Density Estimation ---
density = gaussian_kde(X)   # KDE model
x_vals = np.linspace(X.min(), X.max(), 200)  # Range of values
pdf_vals = density(x_vals)  # Estimated PDF

# Plot PDF
plt.figure(figsize=(8,5))
plt.plot(x_vals, pdf_vals, color="red", label="PDF (KDE)")
sns.histplot(X, bins=10, stat="density", color="lightcoral", edgecolor="black", alpha=0.4)

plt.title("Probability Density Function (PDF) of Sepal Width")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
import pandas as pd

# Load dataset
column_names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
iris = pd.read_csv("iris.data", header=None, names=column_names)

# Threshold
threshold = 4.5

# Probability estimation
prob = (iris['petal_length'] > threshold).mean()

print(f"Probability that petal length > {threshold} cm: {prob:.3f}")
```

```python
import pandas as pd
from scipy.stats import ttest_ind

# Load dataset
column_names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
iris = pd.read_csv("iris.data", header=None, names=column_names)

# Extract petal lengths for two species
setosa = iris[iris['class'] == "Iris-setosa"]["petal_length"]
versicolor = iris[iris['class'] == "Iris-versicolor"]["petal_length"]

# Perform independent t-test
t_stat, p_val = ttest_ind(setosa, versicolor, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_val)

# Interpret results
if p_val < 0.05:
    print(" Significant difference in mean petal length between Setosa and Versicolor")
else:
    print(" No significant difference in mean petal length between Setosa and Versicolor")
```

```python
import pandas as pd

# Load dataset
column_names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
iris = pd.read_csv("iris.data", header=None, names=column_names)

# Extract columns
x = iris["sepal_length"]
y = iris["sepal_width"]

# Covariance matrix
cov_matrix = iris[["sepal_length", "sepal_width"]].cov()
cov_value = cov_matrix.loc["sepal_length", "sepal_width"]

# Correlation matrix
corr_matrix = iris[["sepal_length", "sepal_width"]].corr()
corr_value = corr_matrix.loc["sepal_length", "sepal_width"]

print("Covariance between sepal length and sepal width:", cov_value)
print("Correlation coefficient between sepal length and sepal width:", corr_value)
```
