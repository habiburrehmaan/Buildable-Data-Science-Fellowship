```python
# Step 1: Import NumPy
import numpy as np

# Step 2: Create a 1D array
arr1 = np.array([1, 2, 3, 4, 5])
print("1D Array:")
print(arr1)

# Step 3: Create a 2D array (matrix)
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print("\n2D Array:")
print(arr2)

# Step 4: Basic operations
print("\nAddition (arr1 + 5):")
print(arr1 + 5)

print("\nMultiplication (arr1 * 2):")
print(arr1 * 2)

# Step 5: Array slicing
print("\nSlicing 1D array (arr1[1:4]):")
print(arr1[1:4])   # elements from index 1 to 3

print("\nSlicing 2D array (arr2[0:2, 1:3]):")
print(arr2[0:2, 1:3])   # rows 0-1, cols 1-2
```

```python
# Step 1: Import NumPy
import numpy as np

# Step 2: Create a numerical array
arr = np.array([10, 20, 30, 40, 50, 60])

print("Array:", arr)

# Step 3: Statistical functions
print("\nMean:", np.mean(arr))        # Average value
print("Median:", np.median(arr))      # Middle value
print("Standard Deviation:", np.std(arr))  # Spread of values

print("\nMaximum Value:", np.max(arr))
print("Index of Maximum Value:", np.argmax(arr))  # Position of max value

print("\nMinimum Value:", np.min(arr))
print("Index of Minimum Value:", np.argmin(arr))  # Position of min value
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("cereal.csv")

print("Before Cleaning:")
print(df.isnull().sum())   # Count missing values in each column

# --- Handling Missing Values ---

# 1. Drop rows with missing values
df_drop = df.dropna()
print("\nAfter dropna() - Removed rows with missing values")
print(df_drop.head())

# 2. Fill missing values with a constant (e.g., 0)
df_fill_zero = df.fillna(0)
print("\nAfter fillna(0) - Replaced NaN with 0")
print(df_fill_zero.head())

# 3. Fill missing values with column mean
df_fill_mean = df.fillna(df.mean(numeric_only=True))
print("\nAfter fillna(mean) - Replaced NaN with column mean")
print(df_fill_mean.head())
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv("cereal.csv")

# --- Filtering ---

# Filter cereals with calories greater than 100
high_calories = df[df['calories'] > 100]
print("Cereals with calories > 100:\n", high_calories.head())

# Filter cereals with protein >= 4
high_protein = df[df['protein'] >= 4]
print("\nCereals with protein >= 4:\n", high_protein.head())

# Multiple conditions: calories > 100 AND fat == 1
filtered = df[(df['calories'] > 100) & (df['fat'] == 1)]
print("\nCereals with calories > 100 AND fat == 1:\n", filtered.head())


# --- Sorting ---

# Sort by calories (ascending)
sorted_cal = df.sort_values(by='calories')
print("\nCereals sorted by calories (ascending):\n", sorted_cal.head())

# Sort by calories (descending)
sorted_cal_desc = df.sort_values(by='calories', ascending=False)
print("\nCereals sorted by calories (descending):\n", sorted_cal_desc.head())

# Sort by multiple columns (first protein, then calories)
sorted_multi = df.sort_values(by=['protein', 'calories'], ascending=[False, True])
print("\nCereals sorted by protein (desc) and calories (asc):\n", sorted_multi.head())
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv("cereal.csv")

# --- GroupBy and Aggregations ---

# Average calories per manufacturer
avg_calories = df.groupby('mfr')['calories'].mean()
print("Average calories per manufacturer:\n", avg_calories)

# Count cereals per manufacturer
count_cereals = df.groupby('mfr')['name'].count()
print("\nNumber of cereals per manufacturer:\n", count_cereals)

# Average protein and fat per cereal type
avg_nutrition = df.groupby('type')[['protein', 'fat']].mean()
print("\nAverage protein and fat per cereal type:\n", avg_nutrition)

# Multiple aggregations: min, max, mean of calories by manufacturer
calorie_stats = df.groupby('mfr')['calories'].agg(['min', 'max', 'mean'])
print("\nCalorie stats per manufacturer:\n", calorie_stats)
```

```python
import pandas as pd

# --- Create two DataFrames ---

# Cereal basic info
df1 = pd.DataFrame({
    'cereal_id': [1, 2, 3],
    'name': ['Corn Flakes', 'Honey Nut', 'Choco Crunch'],
    'mfr': ['K', 'G', 'N']
})

# Cereal nutrition info
df2 = pd.DataFrame({
    'cereal_id': [1, 2, 3],
    'calories': [100, 120, 110],
    'protein': [2, 3, 4]
})

print("DataFrame 1:\n", df1)
print("\nDataFrame 2:\n", df2)


# --- Merge Example (like SQL join) ---
merged_df = pd.merge(df1, df2, on='cereal_id')
print("\nMerged DataFrame:\n", merged_df)


# --- Concat Example (stack DataFrames) ---

# Create two small DataFrames with same columns
df_a = pd.DataFrame({
    'name': ['Cereal A', 'Cereal B'],
    'calories': [90, 95]
})

df_b = pd.DataFrame({
    'name': ['Cereal C', 'Cereal D'],
    'calories': [110, 115]
})

# Concatenate row-wise
concat_df = pd.concat([df_a, df_b], ignore_index=True)
print("\nConcatenated DataFrame:\n", concat_df)
```

```python
import matplotlib.pyplot as plt

# Example data
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 40]
y2 = [2, 4, 6, 8, 10]
y3 = [50, 40, 30, 20, 10]
y4 = [1, 3, 5, 7, 9]

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10,8))

# Plot different charts in each subplot
axs[0,0].plot(x, y1, marker='o')
axs[0,0].set_title("Line Plot")

axs[0,1].bar(x, y2, color="orange")
axs[0,1].set_title("Bar Chart")

axs[1,0].scatter(x, y3, color="green")
axs[1,0].set_title("Scatter Plot")

axs[1,1].hist(y4, bins=5, color="purple")
axs[1,1].set_title("Histogram")

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load built-in dataset
tips = sns.load_dataset("tips")

# Set a style
sns.set_style("whitegrid")

# Create a scatter plot with custom palette
sns.scatterplot(x="total_bill", y="tip", hue="time", data=tips, palette="Set2")

# Add title and axis labels
plt.title("Tips vs Total Bill by Time of Day", fontsize=14, fontweight="bold")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip Amount ($)")

plt.show()
```

```python
# Violin plot: tip distribution by day
sns.violinplot(x="day", y="tip", data=tips, palette="muted")

# Add title
plt.title("Distribution of Tips by Day", fontsize=14, fontweight="bold")
plt.xlabel("Day of Week")
plt.ylabel("Tip Amount ($)")

plt.show()
```
